```python
def frequency_chart(dataframe):
    obj_df = dataframe.select_dtypes(include='object')
    for col in obj_df.columns:
        obj_df[col].value_counts().plot(kind='bar')
        plt.title(col)
        plt.show()
        
print(frequency_chart(customer_final))
```

```python
#4a Finding time period of available data
customer_final['tran_date'] = pd.to_datetime(customer_final.tran_date)
```

```python
time_period = customer_final.tran_date.dt.year.max()-customer_final.tran_date.dt.year.min()
print('Time Period in years:', time_period)
```

```python
#4.b Count of transaction where total amount is less than zero
neg_trans = (customer_final.loc[customer_final['total_amt'] < 0])
neg_trans_count = neg_trans.count()['total_amt']
print('neg_trans_count:', neg_trans_count)
```

```python
#5 Analyzing most popular categories among male and female
top_cat_M = customer_final.loc[customer_final.Gender == 'M']
top_cat_M.value_counts('prod_cat').sort_values(ascending = False).head(3)       #top 3 product categories popular among men
```

```python
top_cat_F = customer_final.loc[customer_final.Gender == 'F']
top_cat_F.value_counts('prod_cat').sort_values(ascending = False).head(3)      #TOP 3 CATEGORIES POPULAR AMONG FEMALES
```

```python
#6 City with maximum customers and their percentafe
max_cust = customer_final.value_counts('city_code').head(1) # city with maximum customers
per_of_cust = (max_cust/customer_final.shape[0])*100
print('city with maximum customers:', max_cust , "percentage of city with max customer:", per_of_cust)
```

```python
#7 Store selling maximum products by value and quantity
max_prod_by_value = customer_final.groupby('Store_type')['total_amt'].sum().sort_values(ascending = False).head(1)
max_prod_by_qty = customer_final.groupby('Store_type')['Qty'].sum().sort_values(ascending = False).head(1)
print('max_prod_by_value:',max_prod_by_value)
print('max_prod_by_qty:',max_prod_by_qty)
```

```python
#8 Total amount earned from Electronics and Clothing Categories
flagship_store = customer_final[customer_final.Store_type == 'Flagship store']
flagship_store = flagship_store[flagship_store.prod_cat.isin(['Electronics','Clothing'])]
total_sales = flagship_store.groupby('prod_cat')['total_amt'].sum()
total_sales
```

```python
#9 Total Amount earned from male customers under electronics category
male_cust = customer_final[customer_final.Gender == 'M']
male_electronics = male_cust[male_cust.prod_cat == 'Electronics']
male_electronics.total_amt.sum()
```

```python
#10 customers having more than 10 unique transactions after removing negative values
customer_final = customer_final[customer_final['total_amt'] > 0]
unique_trans = customer_final.groupby('cust_id')['transaction_id'].nunique()
unique_trans = unique_trans[unique_trans > 10]
cust_num = unique_trans.count()
print(cust_num)
```

```python
customer_final['DOB'] = pd.to_datetime(customer_final.DOB)
```

```python
#11a  Total amount spent on electronics and books by customers aging between 25 to 35.
now = dt.datetime.now()
customer_final['Age'] = customer_final['DOB'].apply(lambda x: (now -x).days/365.2425)
customer_final['Age'] = customer_final['Age'].astype(int)
customer_final
```

```python

req_cust = customer_final[(customer_final['Age']> 25) & (customer_final['Age'] < 35)]
req_cust_cat = req_cust[req_cust.prod_cat.isin(["Electronics", 'Books'])]
req_cust_cat.groupby('prod_cat')['total_amt'].sum()
```

```python
#11b Total amount spent by these customers b/w 1st January 2014 to 1st March 2014
date_range = customer_final[(customer_final['tran_date'] > '2014-01-01')&(customer_final['tran_date'] < '2014-03-01')]
total_amt_spent = date_range.total_amt.sum()
total_amt_spent
```
